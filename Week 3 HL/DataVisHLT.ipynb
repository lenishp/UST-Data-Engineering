```python
import altair as alt
import pandas as pd
```

```python
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data",
                    names= ["Sepal length", "Sepal width", "Petal length", "Petal width", "Class"])
```

```python
df.info()
```

```python
chart = alt.Chart(df).mark_point().encode(x = "Sepal length", y = "Sepal width", color = "Class")

chart | chart.encode(x = "Petal length")
```

```python
chart2 = alt.Chart(df).mark_bar().encode(x="Petal width", color = "Class")

chart2
```

```python
inter = alt.selection_interval()
```

```python
chart = alt.Chart(df).mark_point().encode(x = "Sepal length", 
    y = "Sepal width", 
    color = alt.condition(inter,"Class", alt.value("gray"))).add_selection(inter)

chart | chart.encode(x = "Petal length")
```

```python
chart2 = alt.Chart(df).mark_bar().encode(x="Petal width", y = "Class", color = "Class").transform_filter(inter)
```

```python

chart | chart.encode(x = "Petal length",) & chart2
```

I decided to go with this dashboard because from the colour scheme you can easily see the dimensions of sepal and petal(features) of different iris species(class). We can easily separate the setosa species from the other two species as the features are completely varied as we can see from the graph however, the other two species; versicolor and virginica are fairly similar as the sepal length of the two species lies in the similar range. Therefore, I also decided to use box plot as well along with the plot graphs to help me undersatnd the features of the versicolor and virginica more precisely. Using the selection feature on the graph, we can aim a certain region or section within the graph and help us analyse the features of the species more clearly. 

From the dashboard, I can see that the length of sepals of setosa is smaller however the width is bigger in size compared to the other two species. The size of sepals(length and weidth) of both versicolor and virginica is in a similar range and it's harder to distinguish. Eventhough it's harder to distinguish, mostly virginica has bigger sepals as we can see from the range in the graph. The petals follows the similar trend to the size of the sepals; setosa had smaller petals followed by versicolor and then virginica. Unline thesepal, the size of petals are easier to distinguish between versicolor and virginica.
